```python
import re
from pprint import pprint
from numba import njit, types
from numba.typed import Dict

def parser(s):
    lr = None
    maps = Dict.empty(
        key_type=types.unicode_type,
        value_type=types.UniTuple(types.unicode_type, 2)
    )
    for line in s.splitlines():
        if lr is None:
            lr = line
        else:
            if line:
                for node, left, right in re.findall(r"(\w{3}) = \((\w{3}), (\w{3})\)", line, re.I|re.S):
                    maps[node] = (left,right)
    return lr, maps


pprint(parser(readme_input))
```

```python
from itertools import cycle

def part1(lr, maps):
    ptr = 'AAA'
    i=0
    for turn in cycle(lr):
        i+=1
        ptr = maps[ptr][0 if turn=='L' else 1]
        if ptr=='ZZZ':
            return i

print(part1(*parser(readme_input)))
print(part1(*parser(_input)))
```

```python
import math

@njit
def lcm(a, b):
    return abs(a * b) // math.gcd(a, b)

readme_input="""LR

11A = (11B, XXX)
11B = (XXX, 11Z)
11Z = (11B, XXX)
22A = (22B, XXX)
22B = (22C, 22C)
22C = (22Z, 22Z)
22Z = (22B, 22B)
XXX = (XXX, XXX)"""

@njit
def part2(lr, maps):
    ptrs = [x for x in maps if x.endswith('A')]
    mults = []
    for ptr in ptrs:
        i=0
        loop=True
        while loop:
            for turn in lr:
                i+=1
                ptr = maps[ptr][0 if turn=='L' else 1]
                if ptr.endswith('Z'):
                    mults.append(i)
                    loop=False
                    break
    ret = mults[0]
    for n in mults[1::]:
        ret = lcm(ret, n)
    return ret

print(part2(*parser(readme_input)))
print(part2(*parser(_input)))
%timeit -n 10 part2(*parser(_input))
```
