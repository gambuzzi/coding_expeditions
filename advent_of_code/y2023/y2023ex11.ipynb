```python
from pprint import pprint

def expand(s):
    sl = s.splitlines()
    rows = []
    for y, line in enumerate(sl):
        if all(c=='.' for c in line):
            rows.append(y)
    columns = []
    for x in range(len(line)):
        if all(c=='.' for c in (line[x] for line in sl)):
            columns.append(x)
    return columns, rows

def parser(s, exp=1):
    galaxies = set()
    for y,line in enumerate(s.splitlines()):
        for x,cell in enumerate(line):
            if cell=='#':
                galaxies.add((x,y))
    # expansion
    columns, rows = expand(s)
    ret = set()
    for x,y in galaxies:
        # how many empty columns before the x?
        nx = sum(1 for e in columns if e < x)
        # how many empty rows before the y?
        ny = sum(1 for e in rows if e < y)
        ret.add((x+exp*nx, y+exp*ny))

    return ret

pprint(parser(readme_input))
assert parser(readme_input) == {(9, 10), (4, 0), (12, 7), (5, 11), (0, 2), (1, 6), (9, 1), (8, 5), (0, 11)}
```

```python
from itertools import combinations

def part1(g):
    ret = 0
    for a,b in combinations(g, 2):
        ret += abs(a[0]-b[0]) + abs(a[1]-b[1])
    return ret

print(part1(parser(readme_input)))
print(part1(parser(_input)))
```

```python
print(part1(parser(readme_input, 9)))
print(part1(parser(readme_input, 99)))
print(part1(parser(_input, 1_000_000-1)))
```
