```python
from pprint import pprint
from numba import njit

@njit
def parser(s):
    mirrors={}
    splitters={}
    for y, line in enumerate(s.splitlines()):
        for x, c in enumerate(line):
            if c in ('/','\\'):
                mirrors[x,y] = c
            elif c in '-|':
                splitters[x,y] = c
    return mirrors, splitters, x, y



pprint(parser(readme_input))
```

```python
UP = 0,-1
DOWN = 0,1
LEFT = -1,0
RIGHT= 1,0

@njit
def part1(mirrors, splitters, max_x, max_y, start = (-1,0,RIGHT), debug=False):
    REFL = {
        ('\\', UP): LEFT,
        ('\\', LEFT): UP,
        ('\\', DOWN): RIGHT,
        ('\\', RIGHT): DOWN,
        ('/', UP): RIGHT,
        ('/', LEFT): DOWN,
        ('/', DOWN): LEFT,
        ('/', RIGHT): UP,
    }

    seen = set()
    stack = [start]

    def add(x,y,dir):
        if (x,y,dir) not in seen:
            seen.add((x,y,dir))
            stack.append((x,y,dir))

    while stack:
        x,y,dir = stack.pop()
        nx, ny = x+dir[0], y+dir[1]
        if (nx,ny) in mirrors:
            m = mirrors[nx,ny]
            dir = REFL[m, dir]
            add(nx,ny,dir)
        elif (nx,ny) in splitters:
            s = splitters[nx,ny]
            if s=='-' and dir in (UP, DOWN):
                    add(nx,ny,LEFT)
                    add(nx,ny,RIGHT)
            elif s=='|' and dir in (LEFT, RIGHT):
                    add(nx,ny,UP)
                    add(nx,ny,DOWN)
            else:
                if 0<=nx<=max_x and 0<=ny<=max_y:
                    add(nx,ny,dir)
        else:
            if 0<=nx<=max_x and 0<=ny<=max_y:
                add(nx,ny,dir)

    tiles = set([(x,y) for x,y,_ in seen])
    if debug:
        for y in range(max_y+1):
            line = ''
            for x in range(max_x+1):
                line += '.' if (x,y) not in tiles else '#'
            print(line)
    return len(tiles)

print(part1(*parser(readme_input), debug=True))
print(part1(*parser(_input)))

%timeit part1(*parser(_input))
```

```python
def part2(mirrors, splitters, max_x, max_y):
    ret = 0
    for x in range(max_x+1):
        ret = max(ret,
                  part1(mirrors, splitters, max_x, max_y, start = (x,-1,DOWN))
                  )
        ret = max(ret,
                  part1(mirrors, splitters, max_x, max_y, start = (x,max_y+1,UP))
                  )
    for y in range(max_y+1):
        ret = max(ret,
                  part1(mirrors, splitters, max_x, max_y, start = (-1,y,RIGHT))
                  )
        ret = max(ret,
                  part1(mirrors, splitters, max_x, max_y, start = (max_x+1,y,LEFT))
                  )

    return ret

print(part2(*parser(readme_input)))
print(part2(*parser(_input)))

%timeit part2(*parser(_input))
```
