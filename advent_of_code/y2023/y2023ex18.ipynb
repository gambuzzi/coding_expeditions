```python
from pprint import pprint

def parser(s):
    ret = []
    for line in s.splitlines():
        dir, steps, rgb=line.split()
        rgb = rgb.strip('()')
        steps=int(steps)
        ret.append((dir,steps,rgb))
    return ret

pprint(parser(readme_input))
```

```python
delta = {
    'R': (1,0),
    'L': (-1,0),
    'U': (0,-1),
    'D': (0,1),
}

def part1(plan):
    def xy_gen(plan):
        x=y=0
        yield x,y
        for cmd, steps, _ in plan:
            dx,dy = delta[cmd]
            for _ in range(steps):
                x+=dx
                y+=dy
                yield x,y


    grid = set(xy_gen(plan))
    stack = [(1,1)]
    while stack:
        x,y=stack.pop()
        if (x,y) not in grid:
            grid.add((x,y))
            stack.append((x-1,y))
            stack.append((x+1,y))
            stack.append((x,y-1))
            stack.append((x,y+1))
    return len(grid)


print(part1(parser(readme_input)))
print(part1(parser(_input)))

%timeit part1(parser(_input))
```

```python
def part2(plan):
    area = 1
    x=y=0
    min_y=0
    max_y=0
    for _, _, rgb in plan:
        rgb = int(rgb[1:], base=16)
        cmd = 'RDLU'[rgb%16]
        steps = rgb // 16
        dx,dy = delta[cmd]
        # print(cmd, steps)

        if cmd == 'D':
            area += (x+1) * steps
        elif cmd == 'U':
            area -= (x) * steps
        elif cmd == 'R':
            area += steps
        elif cmd == 'L':
            pass

        x+=dx*steps
        y+=dy*steps
        # print(x,',',y,sep='')
        max_y=max(max_y,y)
        min_y=min(min_y,y)

    return area


print(part2(parser(readme_input)))
print(part2(parser(_input)))

%timeit part2(parser(_input))
```

```python
#@title it's matplotlib time! { vertical-output: true}
from IPython.display import HTML, display
from google.colab import files

def progress(value, max=100):
    return HTML("""
        <progress
            value='{value}'
            max='{max}',
            style='width: 100%'
        >
            {value}
        </progress>
    """.format(value=value, max=max))

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation

from matplotlib import rc
rc('animation', html='jshtml')
figsize = 8
fig = plt.figure(figsize=(figsize,figsize))

ax = plt.subplot(1,1,1)

FPS = 4
min_y=max_y=max_x=min_x=0

def gen():
    global min_y, max_y, max_x, min_x
    plan = parser(_input)
    scale = float('+Inf')
    yield []
    to_yield = []
    YIELD = 2000

    for _, _, rgb in plan:
        rgb = int(rgb[1:], base=16)
        cmd = 'RDLU'[rgb%16]
        steps = rgb // 16
        scale = min(scale, steps)

    scale //= 2
    x=y=0
    for _, _, rgb in plan:
        rgb = int(rgb[1:], base=16)
        cmd = 'RDLU'[rgb%16]
        steps = rgb // 16 // scale

        if cmd == 'D':
            for i in range(steps):
                to_yield.append((x, y+i, 10))
                if (len(to_yield)+1) % YIELD==0:
                    yield to_yield
                    to_yield = []
            y+=steps
        elif cmd == 'U':
            for i in range(steps):
                to_yield.append((x, y-i, 0))
                if (len(to_yield)+1) % YIELD==0:
                    yield to_yield
                    to_yield = []
            y-=steps
        elif cmd == 'R':
            x+=steps
            to_yield.append((x, y, 10))
            if (len(to_yield)+1) % YIELD==0:
                yield to_yield
                to_yield = []
        elif cmd == 'L':
            x-=steps

        max_x=max(max_x,x)
        min_x=min(min_x,x)
        max_y=max(max_y,y)
        min_y=min(min_y,y)

    yield to_yield
    for _ in range(FPS*1):
        yield []

data = list(gen())
print(f'len(data) = {len(data)}')

matrix = [[0] * (max_x-min_x+1) for _ in range(max_y-min_y+1)]

def plot(data):
    ax.clear()

    for dx,y,v in data:
        if v>=0:
            for x in range(min_x,dx):
                matrix[y-min_y][x-min_x] = v

    surf = ax.matshow(matrix)
    return surf

out = display(progress(0, len(data)), display_id=True)

def gen2():
    for i, d in enumerate(data):
        yield d
        out.update(progress(i+1, len(data)))

pam_ani = animation.FuncAnimation(fig, plot, frames=gen2, fargs=(), save_count=len(data))
pam_ani.save(
    'y2023ex18.mp4', fps=FPS, extra_args=['-vcodec', 'libx264'])
files.download('y2023ex18.mp4')
# pam_ani.save(
#     'y2023ex18.gif', dpi=300, writer=animation.PillowWriter(fps=FPS))
# files.download('y2023ex18.gif')
# pam_ani
```
