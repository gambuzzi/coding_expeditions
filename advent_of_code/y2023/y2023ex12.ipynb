```python
from pprint import pprint
from numba import njit
from numba.typed import List

@njit
def nint(s):
    ret = 0
    for c in s:
        ret *= 10
        ret += ord(c) - 48
    return ret

@njit
def parser(s):
    ret = List()
    for line in s.splitlines():
        condition1, condition2 = line.split()
        condition2 = List([nint(x) for x in condition2.split(',')])
        ret.append((condition1, condition2))
    return ret

pprint(parser(readme_input))
```

```python
@njit
def all_possible(s):
    ret = List()
    m = list(s)
    idxs = [i for i,x in enumerate(m) if x=='?']
    bits = len(idxs)
    for n in range(2**bits):
        for j in range(bits):
            m[idxs[j]] = '.' if n & 1  else '#'
            n>>=1
        ret.append(''.join(m))
    return ret

@njit
def valid(s, nums):
    s_nums = [0]*0
    pre = None
    cnt = 0
    for c in s+'.':
        if c!='#' and cnt:
            s_nums.append(cnt)
            cnt = 0
        elif c=='#':
            cnt += 1
    if len(nums)==len(s_nums):
        for i in range(len(nums)):
            if nums[i]!=s_nums[i]:
                return False
        return True
    return False

@njit
def part1(data):
    ret = 0
    for c1,c2 in data:
        ret += sum([1 for option in all_possible(c1) if valid(option, c2)])
    return ret

assert list(all_possible('???.###')) == ['###.###', '.##.###', '#.#.###', '..#.###', '##..###', '.#..###', '#...###', '....###']
assert not valid('.##.###', (1,1,3))
assert valid('#.#.###', (1,1,3))

print(part1(parser(readme_input)))
print(part1(parser(_input)))
```

```python
@njit
def p(ss, nums, debug=False):
    ret = 0
    s = ss + '.'
    stack = [(0, nums)]
    while stack:
        start, nums = stack.pop()
        if debug:
            print(start, nums)
        if len(nums)==0 and s.count('#', start)==0:
            ret += 1
        elif len(nums)==0 and s.count('#', start)!=0:
            pass
        elif nums and s.count('#', start)+s.count('?', start)==0:
            pass
        elif len(s)-start < sum(nums)+len(nums)-1:
            pass
        elif s.count('.', start)+s.count('?', start) < len(nums)-1:
            pass
        elif s.count('#', start)+s.count('?', start) < sum(nums):
            pass
        elif s[start]=='.':
            while s[start]=='.' and start<len(s):
                start+=1
            stack.append((start, nums))
        else:
            if s[start]=='?':
                stack.append((start+1, nums))
            if s.count('#', start, start+nums[0])+s.count('?', start, start+nums[0])==nums[0] and s[start+nums[0]] in '.?':
                stack.append((start + nums[0] + 1, nums[1:]))
    return ret

## @njit
def part1b(data):
    ret = 0
    for c1,c2 in data:
        ret += p(c1, c2)
    return ret

print(p('???.###', List([1,1,3])) == 1)
print(p('.??..??...?##.', List([1,1,3])) == 4)
print(p('?#?#?#?#?#?#?#?', List([1,3,1,6])) == 1)
print(p('????.######..#####.', List([1,6,5])) == 4)
print(p('?###????????', List([3,2,1])) == 10)

print(part1b(parser(readme_input)))
print(part1b(parser(_input)))

%timeit -n 2 -r 2 part1b(parser(_input))
```

```python
import functools

s = '..??.????????.???..??.????????.???..'
nums = List([1,3,1,3,1,3])

def q(s, nums):
    l = [x for x in s.split('.') if x]

    @functools.cache
    def fx(ptr_l, ptr_nums):
        if len(l)-ptr_l==1:
            if ptr_nums == len(nums):
                return 1
            return p(l[ptr_l], nums[ptr_nums:])
        if len(l)-ptr_l>1:
            ret = 0
            for n in range(ptr_nums+1, len(nums)+1):
                v = p(l[ptr_l], nums[ptr_nums:n])
                # print('v',v)
                if v==0:
                    break
                ret += v * fx(ptr_l+1, n)
            if '#' not in l[ptr_l]:
                ret += fx(ptr_l+1, ptr_nums)
            return ret
        # print(l,nums)
        return 0

    return fx(0, 0)

def r(ss, nums):
    s = ss + '.'
    @functools.cache
    def fx(ptr_s, ptr_nums, curr_len):
        if ptr_s==len(s) and ptr_nums==len(nums):
            return 1
        if s.count('#', ptr_s)==0 and ptr_nums==len(nums):
            return 1
        if ptr_s>=len(s):
            if ptr_nums>=len(nums):
                return 1
            return 0
        if ptr_nums>=len(nums):
            if s.count('#', ptr_s)>0:
                return 0
            else:
                return 1
        if s[ptr_s]=='.':
            if curr_len==0:
                return fx(ptr_s+1, ptr_nums, 0)
            if curr_len == nums[ptr_nums]:
                return fx(ptr_s+1, ptr_nums+1, 0)
            return 0
        elif s[ptr_s]=='#':
            if curr_len+1>nums[ptr_nums]:
                return 0
            return fx(ptr_s+1, ptr_nums, curr_len+1)
        elif s[ptr_s]=='?':
            # .
            a = 0
            if curr_len==0:
                a = fx(ptr_s+1, ptr_nums, 0)
            if curr_len == nums[ptr_nums]:
                a = fx(ptr_s+1, ptr_nums+1, 0)
            # #
            b = fx(ptr_s+1, ptr_nums, curr_len+1)
            return a + b

    return fx(0,0,0)

# print(p('????????',List([1])))
# print(p('????????',List([1,1])))
# print(p('????????',List([2,1])))
# print(p('????????',List([1,2])))
# print(p('????????',List([1,2,3])))
# print(p('????????',List([2,2])))
# print(p('????????',List([2,3])))
# print(p('????????',List([2,3,3])))

print(p(s,nums))
print(q(s,nums))
print(r(s,nums))
%timeit -n 2 -r 2 p(s,nums)
%timeit -n 2 -r 2 q(s,nums)
%timeit -n 2 -r 2 r(s,nums)

# print(32691779)
# print(q('???????.#??????????.#??????????.#??????????.#??????????.#??', List([1,1,2,1,1,2,1,1,2,1,1,2,1,1,2])))
```
