```python
from pprint import pprint

class Block:
    def __init__(s,rows = None,cols = None):
        s.rows=rows if rows is not None else []
        s.cols=cols if cols is not None else []

    def __repr__(s):
        return f"Block rows={s.rows} cols={s.cols}"

def atoi(s):
    ret = 0
    for c in s:
        ret <<= 1
        if c=='#':
            ret += 1
    return ret


def parser(s):
    blocks = []
    acc = []
    for line in s.splitlines():
        if not line:
            rows = [atoi(x) for x in acc]
            cols = [atoi(x) for x in zip(*acc[::-1])]
            blocks.append(Block(rows, cols))
            acc = []
        else:
            acc.append(line)
    if acc:
        rows = [atoi(x) for x in acc]
        cols = [atoi(x) for x in zip(*acc[::-1])]
        blocks.append(Block(rows, cols))

    return blocks

pprint(parser(readme_input))
```

```python
def find_simm(seq):
    ret = []
    # find 2 elements that are the same
    for i,(a,b) in enumerate(zip(seq, seq[1:])):
        if a==b:  # possible simmetry
            j = 0
            simm = True
            while j+i+1<len(seq) and i-j>=0 and simm:
                if seq[j+i+1] != seq[i-j]:
                    simm = False
                j += 1
            if simm:
                ret.append(i)
    return ret

def part1(blocks):
    ret_rows = 0
    ret_cols = 0
    for block in blocks:
        if i := find_simm(block.rows):
            if len(i)>1:
                raise Exception(f"multiple rows {i}")
            ret_rows += i[0] + 1
        if i := find_simm(block.cols):
            if len(i)>1:
                raise Exception(f"multiple cols {i}")
            ret_cols += i[0] + 1
    # print(ret_rows, ret_cols)
    return 100*ret_rows + ret_cols

print(part1(parser(readme_input)))
print(part1(parser(_input)))

# 31914 low
# 33122 correct

%timeit part1(parser(_input))
```

```python
def bits_off(a,b):
    bits = 0
    while a or b:
        bits += 0 if (a&1)==(b&1) else 1
        a>>=1
        b>>=1
    return bits


def find_dirty_simm(seq):
    ret = set()
    # find 2 elements that are the same
    for start in range(len(seq)-1):
        different_bits = 0
        for i,j in enumerate(range(start,-1,-1)):
            different_bits += bits_off(
                seq[j],
                seq[start+1+i]
            )
            if start+1+i == len(seq)-1:
                break
        if different_bits == 1:
            ret.add(start)
    return ret

def part2(blocks):
    ret_rows = 0
    ret_cols = 0
    for block in blocks:
        if i := find_dirty_simm(block.rows):
            if len(i)>1:
                raise Exception(f"multiple rows {i}")
            ret_rows += i.pop() + 1
        elif i := find_dirty_simm(block.cols):
            if len(i)>1:
                raise Exception(f"multiple cols {i}")
            ret_cols += i.pop() + 1
    # print(ret_rows, ret_cols)
    return 100*ret_rows + ret_cols

print(part2(parser(readme_input)))
print(part2(parser(_input)))

%timeit part2(parser(_input))
```
