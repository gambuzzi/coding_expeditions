```python
from pprint import pprint

class Part:
    def __init__(s,string):
        for p in string.strip('{}').split(','):
            k,v = p.split('=')
            if k in 'xmas':
                s.__dict__[k] = int(v)

    def total_value(s):
        return s.x+s.m+s.a+s.s

    def __repr__(s):
        return f"Part x={s.x} m={s.m} a={s.a} s={s.s}"

    def __getitem__(s, k):
        if k is not None and k in 'xmas':
            return s.__dict__[k]

class Rule:
    def __init__(s,string):
        if ':' in string:
            vcv, dest = string.split(':')
            if '>' in vcv:
                var, value = vcv.split('>')
                condition = '>'
            elif '<':
                var, value = vcv.split('<')
                condition = '<'
            else:
                raise NotImplemented("???")
            s.var = var
            s.condition = condition
            s.value = int(value)
            s.dest = dest
        else:
            s.var = None
            s.condition = None
            s.value = None
            s.dest = string

    def __repr__(s):
        return f"Rule var={s.var} condition={s.condition} value={s.value} dest={s.dest}"

    def match(s, p: Part) -> bool:
        left = p[s.var]
        if s.condition=='<':
            return left<s.value
        if s.condition=='>':
            return left>s.value
        return True

def parser(s):
    parts_section = False
    workflows = {}
    parts = []

    for line in s.splitlines():
        if not line:
            parts_section = True
        elif parts_section:
            parts.append(Part(line))
        else:
            name, rules = line.strip('}').split('{')
            rules = [Rule(x) for x in rules.split(',')]
            workflows[name] = rules
    return parts, workflows
pprint(parser(readme_input))
```

```python
def apply(rules: list[Rule], p: Part) -> str:
    for r in rules:
        if r.match(p):
            return r.dest

def part1(parts, workflows):
    ret = []

    for p in parts:
        next = "in"
        while next is not None and next not in 'AR':
            next = apply(workflows[next], p)
        if next=='A':
            ret.append(p)

    return sum(x.total_value() for x in ret)

print(part1(*parser(readme_input)))
print(part1(*parser(_input)))
```

```python
class Counter:
    def __init__(self, cnt = None):
        if cnt :
            self.x=cnt.x
            self.m=cnt.m
            self.a=cnt.a
            self.s=cnt.s
        else:
            self.x=(1,4000)
            self.m=(1,4000)
            self.a=(1,4000)
            self.s=(1,4000)

    def mul(self):
        ret = self.x[1]-self.x[0]+1
        ret *= self.m[1]-self.m[0]+1
        ret *= self.a[1]-self.a[0]+1
        ret *= self.s[1]-self.s[0]+1
        return ret

    def __getitem__(s, k):
        if k is not None and k in 'xmas':
            return s.__dict__[k]

    def __setitem__(s, k, v):
        if k is not None and k in 'xmas':
            s.__dict__[k] = v

    def __repr__(s):
        return f"Counter x={s.x} m={s.m} a={s.a} s={s.s}"

def part2(workflows: list[Rule]):
    next = "in"
    ret = 0

    # 4000**4 = 256000000000000
    #in
    #px  4000*4000*4000*1350                                                     qqz 4000*4000*4000*(4000-1350)
    #qkq 4000*4000*2006*1350    A 4000*(4000-2090)*(4000-2006)*1350  rfg ...
    cnt = Counter()
    stack = [("in", cnt)]
    while stack:
        next, cnt = stack.pop()
        if next == 'A':
            # print(cnt)
            ret += cnt.mul()
            continue
        if next == 'R':
            continue
        for rule in workflows[next]:
            new_cnt = Counter(cnt)
            next_cnt = Counter(cnt)
            if rule.condition=='<':
                segment = new_cnt[rule.var]
                if segment[0]>=rule.value:
                    continue  # no solution down this branch
                if segment[1]>=rule.value:
                    rest_of_it = rule.value,segment[1]
                    segment = segment[0],rule.value-1
                new_cnt[rule.var] = segment
                next_cnt[rule.var] = rest_of_it
            elif rule.condition=='>':
                segment = new_cnt[rule.var]
                if segment[1]<=rule.value:
                    continue  # no solution down this branch
                if segment[0]<=rule.value:
                    rest_of_it = segment[0], rule.value
                    segment = rule.value+1, segment[1]
                new_cnt[rule.var] = segment
                next_cnt[rule.var] = rest_of_it

            stack.append((rule.dest, new_cnt))
            cnt = next_cnt

    return ret


print(part2(parser(readme_input)[1]), part2(parser(readme_input)[1])==167409079868000)
print(part2(parser(_input)[1]))

%timeit part2(parser(_input)[1])
```
