```python
from pprint import pprint

def parser(s):
    fixed_rocks = {}
    moving_rocks= {}
    for y,line in enumerate(s.splitlines()):
        for x,cell in enumerate(line):
            if cell=='#':
                if x not in fixed_rocks:
                    fixed_rocks[x]=set()
                fixed_rocks[x].add((x,y))
            elif cell=='O':
                if x not in moving_rocks:
                    moving_rocks[x]=set()
                moving_rocks[x].add((x,y))
    return fixed_rocks, moving_rocks, y, x

pprint(parser(readme_input))
```

```python
def north(fixed_rocks, moving_rocks, max_y):
    buckets = {}
    for x in sorted(moving_rocks):
        bucket = [0]
        bucket_start = [0]
        y_ptr = 0
        for y in range(0, max_y+1):
            if (x,y) in moving_rocks[x]:
                bucket[y_ptr]+=1
            elif (x,y) in fixed_rocks.get(x, {}):
                y_ptr += 1
                bucket.append(0)
                bucket_start.append(y+1)
        buckets[x] = bucket_start,bucket
    return buckets

def compute_w(bucket_start,bucket, max_y):
    ret = 0
    for start, count in zip(bucket_start,bucket):
        ten = max_y+1-start
        tot = ten*(ten+1)//2
        minus = (ten-count)*(ten-count+1)//2
        ret += tot-minus
    return ret

def part1(fixed_rocks, moving_rocks, max_y, max_x):
    ret = 0
    for bucket_start,bucket in north(fixed_rocks, moving_rocks, max_y).values():
        ret += compute_w(bucket_start,bucket, max_y)

    return ret

print(part1(*parser(readme_input)))
print(part1(*parser(_input)))
%timeit part1(*parser(_input))
```

```python
from collections import defaultdict
from numba import njit
from numba.typed import List

@njit
def parser_2(s):
    moving_rocks = List([(0,0)]*0)
    sl = s.splitlines()
    Y = len(sl)
    X = len(sl[0])
    fixed_rocks_x = List([List([(0,0)]*0) for _ in range(X)])
    fixed_rocks_y = List([List([(0,0)]*0) for _ in range(Y)])

    for y,line in enumerate(sl):
        for x,cell in enumerate(line):
            if cell=='#':
                fixed_rocks_x[x].append((x,y))
                fixed_rocks_y[y].append((x,y))
            elif cell=='O':
                moving_rocks.append((x,y))
    return fixed_rocks_x, fixed_rocks_y, moving_rocks, y, x

@njit
def compute_w_2(moving_rocks, max_y):
    ret = 0
    for x,y in moving_rocks:
        ret += max_y+1-y
    return ret

@njit
def compute_uuid(moving_rocks, max_y):
    ret = 0
    for x,y in moving_rocks:
        ret += (max_y+1-y)+x*(max_y+1)
    return ret

@njit
def north_2(fixed_rocks_x, moving_rocks, max_y):
    ret = List([(0,0)]*0)
    for (x,y) in moving_rocks:
        blocks = [e for _,e in fixed_rocks_x[x] if e<y]
        ny = max(blocks)+1 if blocks else 0

        while (x,ny) in ret:
            ny+=1
        ret.append((x,ny))
    return ret

@njit
def west_2(fixed_rocks_y, moving_rocks, max_x):
    ret = List([(0,0)]*0)
    for (x,y) in moving_rocks:
        blocks = [e for e,_ in fixed_rocks_y[y] if e<x]
        nx = max(blocks)+1 if blocks else 0

        while (nx,y) in ret:
            nx+=1
        ret.append((nx,y))
    return ret

@njit
def south_2(fixed_rocks_x, moving_rocks, max_y):
    ret = List([(0,0)]*0)
    for (x,y) in moving_rocks:
        blocks = [e for _,e in fixed_rocks_x[x] if e>y]
        ny = min(blocks)-1 if blocks else max_y

        while (x,ny) in ret:
            ny-=1
        ret.append((x,ny))
    return ret

@njit
def east_2(fixed_rocks_y, moving_rocks, max_x):
    ret = List([(0,0)]*0)
    for (x,y) in moving_rocks:
        blocks = [e for e,_ in fixed_rocks_y[y] if e>x]
        nx = min(blocks)-1 if blocks else max_x

        while (nx,y) in ret:
            nx-=1
        ret.append((nx,y))
    return ret

def myprint(fixed_rocks_x, moving_rocks, max_y, max_x):
    ret = [['.']*(max_x+1) for _ in range(max_y+1)]
    for (x,y) in moving_rocks:
        ret[y][x] = 'O'
    for xy in fixed_rocks_x:
        for x,y in xy:
            ret[y][x] = '#'

    print('\n'.join(''.join(e) for e in ret))

@njit
def part2(fixed_rocks_x, fixed_rocks_y, moving_rocks, max_y, max_x):
    seen = {}
    for step in range(1000):
        moving_rocks = north_2(fixed_rocks_x, moving_rocks, max_y)
        moving_rocks = west_2(fixed_rocks_y, moving_rocks, max_x)
        moving_rocks = south_2(fixed_rocks_x, moving_rocks, max_y)
        moving_rocks = east_2(fixed_rocks_y, moving_rocks, max_x)
        # myprint(fixed_rocks_x, moving_rocks, max_y, max_x)
        # print('---')
        uuid = compute_uuid(moving_rocks, max_y)
        if uuid in seen:
            offset = seen[uuid]
            delta = step - seen[uuid]
            # print (seen[uuid], delta, step)
            break
        seen[uuid] = step
        # print(uuid)
    extra_steps = (1000000000 - 1 - offset) % delta
    # print (extra_steps)
    for _ in range(extra_steps):
        moving_rocks = north_2(fixed_rocks_x, moving_rocks, max_y)
        moving_rocks = west_2(fixed_rocks_y, moving_rocks, max_x)
        moving_rocks = south_2(fixed_rocks_x, moving_rocks, max_y)
        moving_rocks = east_2(fixed_rocks_y, moving_rocks, max_x)

    return compute_w_2(moving_rocks, max_y)

print(part2(*parser_2(readme_input)))
print(part2(*parser_2(_input)))

%timeit part2(*parser_2(_input))
```
