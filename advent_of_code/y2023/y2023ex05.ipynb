```python
from pprint import pprint

def parser(s):
    seeds = []
    a_to_b = None
    a_to_b_map = {}
    for line in s.splitlines():
        if line.startswith('seeds:'):
            seeds = list(map(int, line.split()[1:]))
        elif 'map:' in line:
            a_to_b, *_ = line.split()
        elif a_to_b:
            if line:
                destination, source, length = map(int, line.split())
                if a_to_b not in a_to_b_map:
                    a_to_b_map[a_to_b] = []
                a_to_b_map[a_to_b].append((destination, source, length))
    return seeds, a_to_b_map

pprint(parser(readme_input))
```

```python
def part1(seeds, a_to_b):
    def lookup(k, n):
        for (destination, source, length) in a_to_b[k]:
            if source <= n < source + length:
                return destination + (n-source)
        return n
    acc = []
    for seed in seeds:
        n = lookup('seed-to-soil', seed)
        # print(seed,n)
        n = lookup('soil-to-fertilizer', n)
        n = lookup('fertilizer-to-water', n)
        n = lookup('water-to-light', n)
        n = lookup('light-to-temperature', n)
        n = lookup('temperature-to-humidity', n)
        n = lookup('humidity-to-location', n)
        acc.append(n)

    return min(acc)

print(part1(*parser(readme_input)))
print(part1(*parser(_input)))
```

```python
class NumberRange:
    def __init__(self, start=None, end=None):
        if start is not None and end is not None:
            self.range = [(start, end)]
        else:
            self.range = []

    def remove_range(self, start, end):
        new_ranges = []
        for r_start, r_end in self.range:
            # entirely outside the current range
            if end < r_start or start > r_end:
                new_ranges.append((r_start, r_end))
            # entirely covers the current range
            elif start <= r_start and end >= r_end:
                pass
            else:
                # splitting the range into smaller ones
                if start <= r_start:
                    new_ranges.append((end+1, r_end))
                elif end >= r_end:
                    new_ranges.append((r_start, start-1))
                else:
                    new_ranges.append((r_start, start-1))
                    new_ranges.append((end+1, r_end))
        self.range = new_ranges

    def compute_overlap(self, other_start, other_end):
        overlaps = []

        for r_start, r_end in self.range:
            if other_end < r_start or other_start > r_end:
                continue  # no overlap
            else:
                overlap_start = max(other_start, r_start)
                overlap_end = min(other_end, r_end)
                overlaps.append((overlap_start, overlap_end))

        return overlaps


num_range = NumberRange(1, 20)
print("Initial Range:", num_range.range)

num_range.remove_range(5, 30)
print("Range after removing (5, 30):", num_range.range)

num_range = NumberRange(1, 20)
print("Initial Range:", num_range.range)

num_range.remove_range(5, 10)
print("Range after removing (5, 10):", num_range.range)

num_range.remove_range(15, 18)
print("Range after removing (15, 18):", num_range.range)

print("Overlap (10,12):",  num_range.compute_overlap(10,12))
print("Overlap (10,19):",  num_range.compute_overlap(10,19))
```
